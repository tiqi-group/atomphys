```python
import re

pattern = r'^(?P<A>\d*)(?P<element>[A-Za-z]*)(?P<charge>\d*\+|\+*)$'
regex = re.compile(pattern)


def extract_info(input_string):
    match = regex.match(input_string)
    if match:
        A = int(match.group('A')) if match.group('A') else 0
        element = match.group('element')
        charge = match.group('charge')
        if charge:
            num_charge = charge.count('+') if len(set(charge)) == 1 else int(charge[:-1])
        else:
            num_charge = 0
        return A, element, num_charge
    else:
        return None


# Test with example input strings
input_strings = ["13Ca+", "14C", "Yb++", "Be7+"]
for input_string in input_strings:
    info = extract_info(input_string)
    if info:
        A, element, num_charge = info
        print("Input:", input_string)
        print("A:", A)
        print("Element:", element)
        print("Num Charge:", num_charge)
        print()
    else:
        print("No match found.")
```
